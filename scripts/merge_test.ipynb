# Group Merge

```python
import os
from google.colab import userdata
from huggingface_hub import ModelCard, ModelCardData, HfApi, login
import torch
token='yours'
api = HfApi()
# Model pool: change paths if local or hosted elsewhere
user = 'tanganke/'
model_pool = [
    "gpt2_cola",
    "gpt2_mnli",
    "gpt2_mrpc",
    "gpt2_qnli",
    "gpt2_qqp",
    "gpt2_rte",
    "gpt2_sst2"
]

output_dir = "./slerp_outputs"
mask_arg = ""  # Optional; use "" if you don't need it
upload_user = 'yedi-hu/'

os.makedirs(output_dir, exist_ok=True)

# Merge adjacent model pairs
for i in range(0, len(model_pool) - 1, 1):
    model_a = model_pool[i]
    model_b = model_pool[i + 1]

    # Create output name with both parent names
    child_name = f"slerp_{model_a}_{model_b}_{i}"
    output_path = os.path.join(output_dir, child_name)

    command = f"""slerp_merge \\
    {user}{model_a} \\
    {user}{model_b} \\
    {output_path} \\
    {mask_arg}"""

    print(f"\n>>> Merging: {model_a} + {model_b} -> {child_name}")

    !{command}
    repo_name = upload_user+child_name
    try:
      api.create_repo(
          repo_id=repo_name,
          private=False,
          exist_ok=True
        )
    except Exception as e:
        raise Exception(f"Failed to create repository: {str(e)}")

    print(f"Uploading model to {repo_name}...")


    api.upload_folder(
        folder_path=output_path,
        repo_id=repo_name,
        repo_type="model",
        commit_message="Upload model files",
    )
```
